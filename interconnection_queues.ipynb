```python
import pandas as pd
all_data = pd.read_excel('queues_2024.xlsx', sheet_name=1)
all_data
```

```python
all_data.columns
```

```python
def standardize_dates(series):
    return pd.to_datetime(series, 
                          infer_datetime_format=True,
                          errors='coerce')
```

```python
all_data['q_date_standardized'] = standardize_dates(all_data['q_date'])
```

```python
cutoff = pd.Timestamp('2022-01-01')
filtered_data = all_data.loc[all_data['q_date_standardized'] >= cutoff]
filtered_data
```

```python
# verifying that cutoff was correctly implemented
filtered_data.sort_values(by='q_date_standardized')
```

```python
filtered_data['ia_date_standardized'] = standardize_dates(filtered_data['ia_date'])
filtered_data['wd_date_standardized'] = standardize_dates(filtered_data['wd_date'])
filtered_data['on_date_standardized'] = standardize_dates(filtered_data['on_date'])
```

```python
# q_date: date when project entered queue
# ia_date: date of signed interconnection agreement
# wd_date: date project withdrawn from queue
# on_date: date project became operational

selected_cols = ['q_id', 'entity', 'q_date_standardized', 
                 'ia_date_standardized', 'wd_date_standardized', 'on_date_standardized']
```

```python
filtered_data[selected_cols]
```

```python
filtered_data['region'].value_counts()
```

```python
filtered_data['region'] = filtered_data['region'].replace('NYISO', 'ISO-NY')
```

```python
selected_entities = ['PJM', 'ERCOT', 'CAISO', 'SPP', 'ISO-NY', 'ISO-NE']
```

```python
filtered_data = filtered_data[filtered_data['region'].isin(selected_entities)]
filtered_data
```

```python
filtered_data[selected_cols]
```

```python
def queue_duration(df, start, end):
    difference = df[end] - df[start]
    return difference.dt.days
```

```python
filtered_data['ia_duration'] = queue_duration(filtered_data, 'q_date_standardized', 'ia_date_standardized')
filtered_data['wd_duration'] = queue_duration(filtered_data, 'q_date_standardized', 'wd_date_standardized')
filtered_data['on_duration'] = queue_duration(filtered_data, 'q_date_standardized', 'on_date_standardized')
```

```python
selected_cols = selected_cols + ['ia_duration', 'wd_duration', 'on_duration']
selected_cols
```
